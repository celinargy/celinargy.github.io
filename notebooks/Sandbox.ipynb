```python
#libraries
import matplotlib
import matplotlib.pyplot as plt
import squarify # pip install squarify (algorithm for treemap)&lt;/pre&gt;
 
# Create a dataset:
my_values=[i**3 for i in range(1,100)]
 
# create a color palette, mapped to these values
cmap = matplotlib.cm.Blues
mini=min(my_values)
maxi=max(my_values)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in my_values]
 
# Change color
squarify.plot(sizes=my_values, alpha=.8, color=colors )
plt.axis('off')
plt.show()
```

```python
# libraries
import matplotlib.pyplot as plt
import squarify    # pip install squarify (algorithm for treemap)
import pandas as pd
import seaborn as sb

# Data
df = pd.DataFrame({'nb_people':[9.7,35.5,31.3,18.2,5.2,20.7,35.2,44,55.9,44.1], 'group':['18-24 yrs','25-34 yrs', '35-44 yrs','45-44 yrs','55-64 yrs', 'low income','medium income', 'high income', 'male','female'] })
values = df['nb_people'].values.tolist()
# plot
squarify.plot(sizes=df['nb_people'], value=[f'{x:.2f}%' for x in values], color=sb.color_palette("mako", len(values)),label=df['group'], alpha=1, pad=0, text_kwargs={'fontsize':11, 'fontname':"Segoe UI",'color':'white','weight':'bold'} )
plt.axis('off')
plt.show()
```
